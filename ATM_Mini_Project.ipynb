```python
class ATM:
    def __init__(self):
        self.account_number = 123456
        self.pin = 1111
        self.account_type = "saving"
        self.balance = 200000
        self.mobile = 9870865699
        self.menu()
    def menu(self):
        print("""what would you like to do!
        1. account_number
        2. account_type
        3. balance
        4. mobile_number
        5. exit""")
        option = int(input("enter your option: "))
        if option == 1:
            self.bank_account_number()
        elif option == 2:
            self.bank_account_type()
        elif option == 3:
            self.check_balance()
        elif option == 4:
            self.phone_number()
        elif option == 5:
            self.withdraw()
        elif option == 6:
            self.deposit()
        else:
            print("Choose Correct Option")
    def bank_account_number(self):
        input_pin = int(input("enter your pin: "))
        if input_pin == self.pin:
            input_number = int(input("enter your bank account number: "))
            if input_number == self.account_number:
                print("Welcome to SBI Bank ATM")
                print("*"*50)
            else:
                print("""Incorrect Account Number!!! \n'Please Enter Correct Account Number'""")
        else:
            print("incorrect Pin")
        self.menu()
    def bank_account_type(self):
        input_pin = int(input("enter your pin: "))
        if input_pin == self.pin:
            input_account_type = input("Write your Account type: ")
            if input_account_type == self.account_type:
                print("*"*50)
                print("You have Saving Bank Account in SBI")
                print("*"*50)
            else:
                print("please Write correct Type")
            self.menu()
    def check_balance(self):
        input_pin = int(input("enter your pin: "))
        if input_pin == self.pin:
            print("*"*50)
            print(f"your balance is {self.balance}")
            print("*"*50)
        else:
            print("incorrect pin")
        self.menu()
    def phone_number(self):
        input_pin = int(input("enter your pin: "))
        if input_pin == self.pin:
            input_phone_number = int(input("enter your mobile number: "))
            if input_phone_number == self.mobile:
                print("*"*50)
                print("Correct Mobile Number")
                print("*"*50)
            else:
                print("Wrong Mobile Number")
        else:
            print("Wrong Pin")
        self.menu()   
    def withdraw(self):
        input_pin = int(input("enter your pin: "))
        if input_pin == self.pin:
            input_balance = int(input("enter amount: "))
            if  input_balance <= self.balance:
                self.balance = self.balance - input_balance
                print("*"*50)
                print(f"your updated balance is {self.balance}")
                print("*"*50)
            else:
                print("insufficient balance")
        else:
            print("incorrect pin")
        self.menu()
    def deposit(self):
        input_pin = int(input("enter your pin: "))
        if input_pin == self.pin:
            input_deposit = int(input("enter amount: "))
            self.balance = self.balance + input_deposit
            print("*"*50)
            print(f"your updated balance is {self.balance}")
            print("*"*50)
        else:
            ("incorrect pin")
        self.menu()
 
sbi = ATM()
```
